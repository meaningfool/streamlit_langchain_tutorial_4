```python
with open('examples/state_of_the_union.txt', 'r') as f:  # If myfile.txt is a text file
    content1 = [f.read()]
```

```python
from langchain.text_splitter import TokenTextSplitter

text_splitter = TokenTextSplitter(chunk_size= 1000, chunk_overlap=0)
chunks = text_splitter.create_documents(content1)
```
